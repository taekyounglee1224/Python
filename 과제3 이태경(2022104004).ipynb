```python
#Class 1.
#fourcalculation.py
"""A class for four arithmetic calculations"""

class FourCalculation:
    """A class for four arithmetic caluclations"""
    
    def __init__(self,first,second):
        self.first=first
        self.second=second
        #if second is zero, raise an exception
        if second==0:
            raise ValueError('The second number must not be 0')
            
    def add(self):
        return self.first+self.second
    
    def mul(self):
        return self.first*self.second
    
    def sub(self):
        return self.first-self.second
    
    def div(self):
        return self.first/self.second
    
```

```python
b=FourCalculation(8,2)
print(b.first,b.second)
print(b.add(),b.mul(), b.sub(),b.div())
```

```python
#Class 2.
class Person:
    def __init__(self, first, last):
        self.firstname=first
        self.lastname=last
        
    def Name(self):
        return self.firstname+" "+self.lastname
    
class Student(Person):
    def __init__(self, first, last, studentnum):
        Person. __init__(self, first, last) 
        self.studentnumber=studentnum
    
    def Getstudent(self):
        return self.Name()+ "," +self. studentnumber
    
```

```python
x=Person("TaeKyoung", "Lee")
print(x.Name())
y=Student("TaeKyoung", "Lee", "2022104004")
print(y.Getstudent())
```

```python
#Class 3.
class AreaVolume:
  def __init__(self, radius, height):
    self.radius=radius
    self.height=height
  
  def circle(self):
    return 3.14*(self.radius)**2
  
  
  def cyl(self):
    return 3.14* (self.radius)**2*(self.height)

  def cone(self):
    return (1/3)*3.14*(self.radius)**2*(self.height)

  
```

```python
c=AreaVolume(4,5)
print(c.radius,c.height)
print(f'원의 넓이는 {c.circle()}, 원기둥의 부피는 {c.cyl():.2f}, 원뿔의 부피는 {c.cone():.2f}입니다')
```

```python
#Class 4.
class PhoneNum:
  def __init__(self,name,phnum):
    self.name= name
    self.phnum= phnum
  
  def Address(self):
    return self.name+" "+self.phnum

class Identity(PhoneNum):
  def __init__(self,name,phnum,regnum):
    PhoneNum.__init__(self, name, phnum)
    self.regnum=regnum
  
  def Getidentity(self):
    return self.Address()+" "+self.regnum
```

```python
d=PhoneNum("TaeKyoungLee", "01072720913")
print(d.Address())
e=Identity("TaeKyoungLee","01072720913", "0112243077310")
print(e.Getidentity())
```

```python
#Class 5.
class Rectangle:
 
    def __init__(self, width, height):
        self.width = width
        self.height = height
 
    def calcArea(self):
        area = self.width * self.height
        return area
```

```python
f=Rectangle(10,8)
print(f.calcArea())
```
