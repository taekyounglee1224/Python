```python
#Q1. 다음 코드의 출력값은 무엇인가?
a= "Life is too short, you need Python"

if "wife" in a:
  print("wife")
elif "Python" in a and "you" not in a:
  print("Python")
elif "shirt" not in a:
  print("shirt")
elif "need" in a:
  print("need")
else:
  print("None")
```

```python
#Q2. 'while'문을 사용해서 1부터 1000까지 자연수 중 3의 배수의 합을 구해보자.
i=1
result=0
while i<=1000:
  if i%3==0:
    result+=i
  i+=1
print(result)
```

```python
#Q5. 10명의 학생의 평균점수를 구해보자.
grades=[70,55,60,75,95,90,80,80,85,100]
total=0
numstudents=0
for i in grades:
  total=total+i 
  numstudents+=1
average=total/numstudents
print(average)
```

```python
#본인 Q2. 
height=float(input('Enter your height in meters'))
weight=float(input('Enter your weight in kg'))
BMI=weight/height**2
if BMI>=30:
    print(f'Your BMI is {BMI:.2f}','Severely Obese')
elif BMI>=25:
    print(f'Your BMI is {BMI:.2f}','Obese')
elif BMI>=23.5:
    print(f'Your BMI is {BMI:.2f}','Overweight')
elif BMI>=18.5:
    print(f'Your BMI is {BMI:.2f}','Normal')
else:
    print(f'Your BMI is {BMI:.2f}','Underweight')
```

```python
#본인 Q4. 2부터 100까지 중 소수만 모두 출력하라
for i in range(2,101):
  a=True
  for j in range(2,i):
    if i%j==0:
      a=False
      break
  if a:
      print(i, end=" ")

```

```python
#본인 Q5. 구구단 표를 출력하라
for i in range(2,10):
  for j in range(1,10):
    print(i,"x",j,"=",i*j, end="    ")
  print()
```

```python
#본인 Q6. 다음 학생들의 키 중 키가 170 이상인 학생들은 몇명인가?
height=[167,180,178,165,171,182,168,177,190,184]
numstudents=0
for x in height:
  if x>=170:
    numstudents+=1
print(f'키가 170보다 큰 학생들의 수는 {numstudents}명 입니다.')
```
