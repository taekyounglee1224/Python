```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('apt.csv',encoding='cp949')
```

```python
df
```

```python
#분석1. 건축년도에 따른 건물의 층수
plt.scatter(df['건축년도'],df['층'])
plt.xlabel('Built Year',fontsize=10)
plt.ylabel('Floors',fontsize=10)
plt.show()
#분석 결과: 최근에 지어진 건물일수록 고층 건물이 많다.
```

```python
df1=df.loc[:, ['지역','건축년도','층','면적','가격']] [df.지역=='서울특별시 서초구 서초동'] [df.건축년도>1995] [df.층<25]
df1
```

```python
#분석2: 서울시 서초구 서초동에 1995년도 이후 지어진 아파트 중 25층 이하 아파트의 총 면적에 따른 매매가격 추이 
df1.figsize=(10,10)
plt.scatter(df1['면적']*df1['층'],df1['가격'])
plt.xlabel('Total Area',fontsize=10)
plt.ylabel('Price',fontsize=10)
plt.show()
#분석결과: 총 면적(면적*층수)가 클 수록 가격이 올라간다.
```

```python
df['단가']=df.가격/df.면적
df2=df.loc[5720:6653, ['지역','가격','면적','단가','건축년도']] [df.건축년도>2000]
df2
```

```python
#분석3: 경기도 수원영통구에 지어진 아파트 중 2000년도 이후에 지어진 아파트들의 건축년도에 따른 단가 추이
plt.scatter(df2['건축년도'],df2['단가'])
plt.xlabel('Built year',fontsize=10)
plt.ylabel('Unit Price',fontsize=10)
plt.show()
#분석 결과: 최근에 지어진 아파트일수록 전체적으로 단가가 높아지는 추세
            #대략 2007년도를 기점으로 급격하게 단가가 증가하였고, 2011~2014년 사이에 단가가 가장 높았다.
```
