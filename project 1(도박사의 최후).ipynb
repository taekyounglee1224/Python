```python
#Project: 주식투자자의 고민 
#원금 10,000$의 주식투자자는 주식거래를 얼마나 하는 것이 합리적일까?
import random
import matplotlib.pyplot as plt

x=10000
z=list()
i=0

while x>0 and i<20000:
    y=random.sample(list(range(-100,100)),k=1)
    x=x+y[0]
    z.append(x)
    i=i+1

plt.scatter(list(range(i)),z)
plt.xlim(0,); plt.ylim(0,)
plt.title("The end of the gambler")
plt.xlabel("Number of gambling",labelpad=10)
plt.ylabel("Money", labelpad=10)
plt.show()
```
