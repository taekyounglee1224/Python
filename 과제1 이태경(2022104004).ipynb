```python
#Q1. 홍길동 씨의 과목별 점수는 다음과 같다. 홍길동 씨의 평균 점수를 구해보자.
total=0
numsubjects=0
grades=[80,75,55]
for i in grades: 
  total+=i
  numsubjects+=1
  average=total/numsubjects
print(f'홍길동 씨의 평균 점수는 {average}입니다')
```
